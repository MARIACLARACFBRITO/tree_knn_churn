To use the data in Machine Learning algorithms, we need to define the explanatory variables and the target variable. In this challenge, separate the churn dataset into explanatory variables, storing them in a variable X, and the target variable in y.

````python
x=dados.drop('churn',axis=1)
y=dados['churn']
````

````python
x
````

````python
y
````

Categorical variables in text format cannot be directly used in Machine Learning models. In this challenge, transform the categorical variables into numerical format using OneHotEncoder, with the parameter drop='if_binary' if any variable has only 2 categories.

````python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
````

````python
colunas=x.columns
````

````python
dados.info()
````

````python
one_hot=make_column_transformer((
    OneHotEncoder(drop='if_binary'),
    ['pais','sexo_biologico']
),
remainder='passthrough' ,
sparse_threshold=0)
````

````python
x=one_hot.fit_transform(x)
````

````python
one_hot.get_feature_names_out(colunas)
````

````python
pd.DataFrame(x,columns=one_hot.get_feature_names_out(colunas))
````

The target variable, being categorical, also requires similar processing as the categorical explanatory variables to be used in algorithms. In this task, use the LabelEncoder method to transform the churn variable.

````python
from sklearn.preprocessing import LabelEncoder
````

````python
label_encoder=LabelEncoder()
````

````python
y=label_encoder.fit_transform(y)
````

````python
y
````

Splitting the data into training and test sets is essential to understand if a model can learn patterns and generalize to new data. In this task, perform a stratified split of the dataset into training and test sets.

````python
from sklearn.model_selection import train_test_split
````

````python
x_treino,x_teste,y_treino,y_teste=train_test_split(x,y,stratify=y,random_state=5)
````

A baseline model is crucial for establishing a comparison criterion for more complex models. In this step, create a baseline model using the DummyClassifier and find the accuracy using the score method.

````python
from sklearn.dummy import DummyClassifier
````

````python
dummy=DummyClassifier()
dummy.fit(x_treino,y_treino)

dummy.score(x_teste,y_teste)
````

The decision tree is an algorithm that makes classifications based on simple decisions derived from the data. We need to be cautious not to use too large a depth, as this can lead to overfitting the model to the training data. In this challenge, create a decision tree model with the parameter max_depth=4, evaluate the model's performance on the test data, and visualize the tree's decisions using the plot_tree method.

````python
from sklearn.tree import DecisionTreeClassifier
````

````python
arvore=DecisionTreeClassifier(max_depth=4)
arvore.fit(x_treino,y_treino)
````

````python
arvore.predict(x_teste)
````

````python
arvore.score(x_teste,y_teste)
````

````python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
````

````python
nome_colunas=['Alemanha','Espanha','França','Mulher','Homem','credito','idade','anos cliente','saldo','serviços adquiridos','tem cartao crédito','membro ativo','salario estimado']
````

````python
plt.figure(figsize=(10,6))
plot_tree(arvore, filled=True, class_names=['nao','sim'], fontsize=5, feature_names=nome_colunas);
````

Data normalization is an important task to keep all numerical values on the same scale and ensure that all features have an equal impact on the model. In this task, normalize the dataset using the MinMaxScaler.

````python
from sklearn.preprocessing import MinMaxScaler
````

````python
normalizacao=MinMaxScaler()
x_treino_normalizado=normalizacao.fit_transform(x_treino)
````

````python
pd.DataFrame(x_treino_normalizado)
````


With the data normalized, we can use the KNN model, which calculates distances to find the nearest neighbors. In this activity, create a KNN model using the KNeighborsClassifier with the normalized data, and evaluate its performance on the normalized test data as well.

````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
knn=KNeighborsClassifier()
knn.fit(x_treino_normalizado,y_treino)
````

````python
x_teste_normalizado=normalizacao.transform(x_teste)
````

````python
knn.score(x_teste_normalizado,y_teste)
````
